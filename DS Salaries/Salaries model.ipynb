```python
df = pd.read_csv('ds_salaries.csv')
table1 = df.groupby(['job_title','experience_level']).mean().round()
table1 = table1.drop(columns=['Unnamed: 0','salary'])
table1.sort_values(by=['salary_in_usd'],ascending=False)
table1.head(50)
```
